```python
from huggingface_hub import HfApi, upload_file
from getpass import getpass
import os

HF_USERNAME = "zotthytt12"
REPO_NAME = "model_hr"
HF_REPO_ID = f"{HF_USERNAME}/{REPO_NAME}"

HF_TOKEN = getpass("Wprowadź swój token Hugging Face (write): ")

model_path = "./model_raport.pkl" 

if not os.path.exists(model_path):
    print(f"BŁĄD: Nie znaleziono pliku modelu w ścieżce: {model_path}")
else:
    print(f"Znaleziono model: {model_path}. Rozpoczynam wysyłanie...")
    
    api = HfApi()

    print(f"Tworzenie/sprawdzanie repozytorium: {HF_REPO_ID}")
    api.create_repo(repo_id=HF_REPO_ID, private=False, exist_ok=True, token=HF_TOKEN)

    print("Wrzucanie pliku modelu...")
    upload_file(
        path_or_fileobj=model_path, 
        path_in_repo="model_raport.pkl",
        repo_id=HF_REPO_ID, 
        token=HF_TOKEN
    )

    print("Tworzenie i wrzucanie pliku README.md...")
    readme_content = f"# {REPO_NAME}\n\nModel do klasyfikacji cv na podstawie danych HR.\n"
    
    with open("README.md", "w", encoding="utf-8") as f:
        f.write(readme_content)
        
    upload_file(
        path_or_fileobj="README.md", 
        path_in_repo="README.md", 
        repo_id=HF_REPO_ID, 
        token=HF_TOKEN
    )
    
    os.remove("README.md")

    print("\n--- Gotowe! ---")
    print(f"Model i README zostały wysłane do: https://huggingface.co/{HF_REPO_ID}")
```
